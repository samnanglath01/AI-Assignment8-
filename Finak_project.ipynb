```python
##DATA PREPROCESSInG


# Function to impute missing values
from sklearn.compose import ColumnTransformer
from sklearn.discriminant_analysis import StandardScaler
from sklearn.preprocessing import OneHotEncoder


def fill_missing_values(df):
    for column in df.columns:
        if df[column].dtype in ['float64', 'int64']:
            df[column] = df[column].fillna(df[column].median())
        elif df[column].dtype == 'object':
            df[column] = df[column].fillna(df[column].mode()[0])
    return df
df = fill_missing_values(df)
print(df.isnull().sum())
if df.isnull().sum().any():
    print("There are still missing values in the dataset.")
else:
    print("No missing values found.")
X = df.drop(columns=['RainTomorrow'])  # Drop the target column
y = df['RainTomorrow'] 
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Create transformers
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

# Column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])


# Apply transformations
X_preprocessed = preprocessor.fit_transform(X)


    

```

```python

#encode target
from sklearn.preprocessing import LabelEncoder

# Create label encoder object
label_encoder = LabelEncoder()

# Fit and transform the target variable
y_encoded = label_encoder.fit_transform(y)

# Checking the mapping
print(dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
```

```python
# Split the data into training and testing sets, ensuring balanced classes
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Initialize classifiers
decision_tree = DecisionTreeClassifier(random_state=42)
svm = make_pipeline(StandardScaler(with_mean=False), SVC(random_state=42))
mlp = make_pipeline(StandardScaler(with_mean=False), MLPClassifier(random_state=42))
logistic_regression = make_pipeline(StandardScaler(with_mean=False), LogisticRegression(random_state=42))
naive_bayes = GaussianNB()
bagging = BaggingClassifier(DecisionTreeClassifier(random_state=42), n_estimators=10, random_state=42)

# Dictionary to hold classifiers
classifiers = {
    'Decision Tree': decision_tree,
    'SVM': svm,
    'MLP': mlp,
    'Logistic Regression': logistic_regression,
    'Naive Bayes': naive_bayes,
    'Bagging': bagging
}

# Ensure X_train is dense for models that don't support sparse matrices
if hasattr(X_train, "toarray"):  # Check if X_train is sparse
    X_train_dense = X_train.toarray()
else:
    X_train_dense = X_train

# Train each model and evaluate on the testing set
for name, clf in classifiers.items():
    try:
        # Check if the classifier can handle sparse data, use dense if not
        if 'Naive Bayes' in name:
            clf.fit(X_train_dense, y_train)
        else:
            clf.fit(X_train, y_train)

        # Predict on testing data
        if hasattr(X_test, "toarray"):
            y_pred = clf.predict(X_test.toarray())
        else:
            y_pred = clf.predict(X_test)

        # Evaluate accuracy
        accuracy = accuracy_score(y_test, y_pred)
        print(f'Accuracy of {name}: {accuracy:.2f}')
    except Exception as e:
        print(f'Error training {name}: {e}')
```

```python

from sklearn.model_selection import GridSearchCV

# Decision Tree Grid Search
param_grid_dt = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5, 10]
}
grid_search_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=5, scoring='accuracy')
grid_search_dt.fit(X_train, y_train)
print("Best Decision Tree Accuracy:", grid_search_dt.best_score_)
print("Best parameters:", grid_search_dt.best_params_)
# SVM Grid Search
param_grid_svm = {
    'svc__C': [0.1, 1, 10, 100],
    'svc__kernel': ['linear', 'rbf']
}
svm_pipeline = make_pipeline(StandardScaler(with_mean=False), SVC(random_state=42))
grid_search_svm = GridSearchCV(svm_pipeline, param_grid_svm, cv=5, scoring='accuracy')
grid_search_svm.fit(X_train, y_train)
print("Best SVM Accuracy:", grid_search_svm.best_score_)
print("Best parameters:", grid_search_svm.best_params_)
# MLP Grid Search
from sklearn.neural_network import MLPClassifier

param_grid_mlp = {
    'mlpclassifier__alpha': [0.0001, 0.001, 0.01],
    'mlpclassifier__hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'mlpclassifier__activation': ['tanh', 'relu']
}
mlp_pipeline = make_pipeline(StandardScaler(with_mean=False), MLPClassifier(random_state=42))
grid_search_mlp = GridSearchCV(mlp_pipeline, param_grid_mlp, cv=5, scoring='accuracy')
grid_search_mlp.fit(X_train, y_train)
print("Best MLP Accuracy:", grid_search_mlp.best_score_)
print("Best parameters:", grid_search_mlp.best_params_)
# Logistic Regression Grid Search
param_grid_lr = {
    'logisticregression__C': [0.01, 0.1, 1, 10, 100],
    'logisticregression__penalty': ['l1', 'l2'],
    'logisticregression__solver': ['liblinear']  # liblinear supports both l1 and l2
}
lr_pipeline = make_pipeline(StandardScaler(with_mean=False), LogisticRegression(random_state=42))
grid_search_lr = GridSearchCV(lr_pipeline, param_grid_lr, cv=5, scoring='accuracy')
grid_search_lr.fit(X_train, y_train)
print("Best Logistic Regression Accuracy:", grid_search_lr.best_score_)
print("Best parameters:", grid_search_lr.best_params_)
# Bagging Grid Search
from sklearn.ensemble import BaggingClassifier

param_grid_bagging = {
    'n_estimators': [10, 50, 100],
    'base_estimator__max_depth': [3, 5, 10]
}
bagging = BaggingClassifier(DecisionTreeClassifier(random_state=42), random_state=42)
grid_search_bagging = GridSearchCV(bagging, param_grid_bagging, cv=5, scoring='accuracy')
grid_search_bagging.fit(X_train, y_train)
print("Best Bagging Accuracy:", grid_search_bagging.best_score_)
print("Best parameters:", grid_search_bagging.best_params_)
```
